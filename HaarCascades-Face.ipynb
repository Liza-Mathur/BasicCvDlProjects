```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
face = cv2.imread('DATA/face.jpg')
face_plt = cv2.cvtColor(face, cv2.COLOR_BGR2RGB)
plt.imshow(face_plt)
```

```python
face_cascade = cv2.CascadeClassifier('DATA/HaarCascades/haarcascade_frontalface_default.xml')
```

```python
img = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
face = face_cascade.detectMultiScale(img, 1.3, 5)
face
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(face_plt, (x,y), (x+w, y+h), (255,0,0), 5)
plt.imshow(face_plt)
```

```python
eye_cascade = cv2.CascadeClassifier('DATA/HaarCascades/haarcascade_eye.xml')
eyes = eye_cascade.detectMultiScale(img , 1.1, 3)
for (x,y,w,h) in eyes:
    cv2.rectangle(face_plt, (x,y), (x+w, y+h), (255,0,0), 5)
plt.imshow(face_plt)
```
