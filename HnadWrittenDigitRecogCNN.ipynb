```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers, models
```

```python
df_train = pd.read_csv('../YTCodeAlongs/Data/mnist_train.csv')
df_train
```

```python
df_test = pd.read_csv('../YTCodeAlongs/Data/mnist_test.csv')
df_test
```

```python
x_train = df_train.iloc[:, 1:].to_numpy().reshape(-1, 28, 28)  
y_train = df_train.iloc[:, 0].to_numpy()  # Assuming first column is the label
```

```python
x_train
```

```python
x_test = df_test.iloc[:, 1:].to_numpy().reshape(-1, 28, 28)  
y_test = df_test.iloc[:, 0].to_numpy()  # Assuming first column is the label
```

```python
x_train[0].shape
```

```python
ann = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(3000, activation='relu'),
    layers.Dense(1000, activation='relu'),
    layers.Dense(10, activation='softmax')
])
ann.compile(optimizer='SGD',
            loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
ann.fit(x_train, y_train, epochs=10)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
y_pred = ann.predict(x_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```

```python
cnn = models.Sequential([
    layers.Conv2D(input_shape = (28,28,1), activation = 'relu', filters= 32, kernel_size=3),
    layers.MaxPooling2D(2),
    layers.Conv2D(activation = 'relu', filters= 64, kernel_size=3),
    layers.MaxPooling2D(2),

    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation = 'softmax')
])

cnn.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

x_train = x_train.reshape(-1, 28, 28, 1)  # Ensure shape matches input_shape
x_test = x_test.reshape(-1, 28, 28, 1)  # Reshape test data similarly

cnn.fit(x_train, y_train, epochs=10)
```

```python
cnn.evaluate(x_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix , classification_report
y_pred = cnn.predict(x_test)
y_pred_classes = [np.argmax(element) for element in y_pred]

print("Classification Report: \n", classification_report(y_test, y_pred_classes))
```
