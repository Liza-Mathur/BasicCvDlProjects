```python
import cv2

# Initialize global variables
draw = False
x1, y1 = -1, -1
img = cv2.imread('DATA/house2.jpg')

# Mouse callback function
def draw_rect(event, x, y, flags, param):
    global x1, y1, draw

    if event == cv2.EVENT_LBUTTONDOWN:  # Mouse button pressed
        draw = True
        x1, y1 = x, y

    elif event == cv2.EVENT_MOUSEMOVE:  # Mouse moves while button is pressed
        if draw:
            img_copy = img.copy()  # Create a fresh copy for live drawing
            cv2.rectangle(img_copy, (x1, y1), (x, y), (0, 255, 0), 2)
            cv2.imshow("Show", img_copy)

    elif event == cv2.EVENT_LBUTTONUP:  # Mouse button released
        draw = False
        cv2.rectangle(img, (x1, y1), (x, y), (0, 255, 0), 2)  # Draw final rectangle
        cv2.imshow("Show", img)

# Create window and set callback
cv2.namedWindow("Show")
cv2.setMouseCallback("Show", draw_rect)

# Main loop
cv2.imshow("Show", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
