```python
import cv2
import matplotlib.pyplot as plt
```

```python
group_img = cv2.imread('../DATA/avenger_crew_group.jpg')
group_img = cv2.cvtColor(group_img, cv2.COLOR_BGR2GRAY)
plt.imshow(group_img)
plt.show()
```

```python
single_person = cv2.imread('../DATA/tony_stark.jpg')
single_person = cv2.cvtColor(single_person , cv2.COLOR_BGR2GRAY)
plt.imshow(single_person, cmap='gray')
```

```python
plt.imshow(group_img)
plt.imshow(single_person, cmap='gray')
```

```python
face_cascade = cv2.CascadeClassifier('../DATA/HaarCascades/haarcascade_frontalface_default.xml')
faces = face_cascade.detectMultiScale(group_img, 1.3 , 5)
faces
```

```python
x , y, w, h = faces[2]
roi = group_img[y:y+h , x:x+w]
plt.imshow(roi , cmap='gray')
```

```python
import os
output_folder = "StoredFaces"
```

```python
for i , (x,y,w,h) in enumerate(faces):
    roi = group_img[max(y-5 , y):max(y+h+5 , y+h) , max(x , x-5):max(x+w, x+w+5)]
    face_path = os.path.join(output_folder, f"face_{i}.jpg")
    cv2.imwrite(face_path, roi)
    print(face_path)
```
